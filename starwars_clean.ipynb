```python
series = []
id_all_global = data_star.columns[3:9]
yes_no_dict = {0:'Star Wars: Episode I The Phantom Menace', 1:'Star Wars: Episode II Attack of the Clones', 2:'Star Wars: Episode III Revenge of the Sith', 3: 'Star Wars: Episode IV A New Hope', 4: 'Star Wars: Episode V The Empire Strikes Back', 5: 'Star Wars: Episode VI Return of the Jedi'}
# series = [series_1, series_2,series_3,series_4,series_5]
for i in id_all_global:
    id = data_star[i]
    for a in range(5):
        yes_no = {
                yes_no_dict[a] : True,
                np.NaN : False
            }
        if i == id_all_global[0]:
            series_1 = id.map(yes_no)
            series_1 = series_1.rename({id_all_global[0] : 'seen_1'})
            series.append(series_1)
        elif i == id_all_global[1]:
            series_2 = id.map(yes_no)
            series_1 = series_1.rename({ id_all_global[1] : 'seen_2'})
            series.append(series_2)
        elif i == id_all_global[2]:
            series_3 = id.map(yes_no)
            series_1 = series_1.rename({ id_all_global[2] : 'seen_3'})
            series.append(series_3)
        elif i == id_all_global[3]:
            series_4 = id.map(yes_no)
            series_1 = series_1.rename({ id_all_global[3] : 'seen_4'})
            series.append(series_4)
        elif i == id_all_global[4]:
            series_5 = id.map(yes_no)
            series_1 = series_1.rename({ id_all_global[4] : 'seen_5'})
            series.append(series_5)
```
